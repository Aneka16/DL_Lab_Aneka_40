```python
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.models import Model
import numpy as np
```

```python
input_dim = 20
latent_dim = 2
```

```python
inputs = Input(shape=(input_dim,))
encoded = Dense(10, activation='relu')(inputs)
latent = Dense(latent_dim)(encoded)
```

```python
decoded = Dense(10, activation='relu')(latent)
outputs = Dense(input_dim, activation='sigmoid')(decoded)
```

```python
vae = Model(inputs, outputs)
vae.compile(optimizer='adam', loss='mse')
```

```python
X = np.random.rand(1000, input_dim)
```

```python
print("\nTraining VAE...")
vae.fit(X, X, epochs=5, batch_size=32)
```

```python
# =====================================================
# 🔹 2. Generative Adversarial Network (GAN)
# =====================================================

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
# Generator
generator = Sequential([
    Dense(16, activation='relu', input_dim=10),
    Dense(1, activation='sigmoid')
])
```

```python
discriminator = Sequential([
    Dense(16, activation='relu', input_dim=1),
    Dense(1, activation='sigmoid')
])
```

```python
discriminator.compile(optimizer='adam', loss='binary_crossentropy')
```

```python
# GAN (combine both)
discriminator.trainable = False

gan = Sequential([generator, discriminator])
gan.compile(optimizer='adam', loss='binary_crossentropy')
```

```python
print("\nTraining GAN...")
```

```python
for epoch in range(3):
    # Generate fake data
    noise = np.random.rand(32, 10)
    fake_data = generator.predict(noise)

    # Real data
    real_data = np.random.rand(32, 1)

    # Labels
    real_labels = np.ones((32, 1))
    fake_labels = np.zeros((32, 1))

    # Train discriminator
    discriminator.train_on_batch(real_data, real_labels)
    discriminator.train_on_batch(fake_data, fake_labels)

    # Train generator via GAN
    noise = np.random.rand(32, 10)
    gan.train_on_batch(noise, real_labels)

    print(f"Epoch {epoch+1} completed")
```

```python
# =====================================================
# 🔹 3. Graph Convolutional Network (GCN)
# =====================================================

import torch
import torch.nn as nn
```

```python
class SimpleGCN(nn.Module):
    def __init__(self):
        super().__init__()
        self.fc = nn.Linear(10, 2)

    def forward(self, x, adj):
        x = torch.matmul(adj, x)   # Graph aggregation
        x = self.fc(x)
        return x
```

```python
x = torch.rand(5, 10)   # 5 nodes, 10 features
adj = torch.eye(5)      # Identity matrix (simple graph)

model = SimpleGCN()

output = model(x, adj)

print("\nGCN Output:")
print(output)
```
