```python
import torch
import torch.nn.functional as F
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sentence = ["I", "love", "deep", "learning"]
```

```python
attention_weights = torch.tensor([0.1, 0.3, 0.4, 0.2])
```

```python
attention_weights = F.softmax(attention_weights, dim=0).detach().numpy()

print("Normalized Attention Weights:", attention_weights)
```

```python
plt.figure(figsize=(8, 2))

sns.heatmap(
    [attention_weights],
    annot=True,
    xticklabels=sentence,
    yticklabels=["Attention"],
    cmap="Blues"
)

plt.title("Attention Heatmap")
plt.xlabel("Words")
plt.ylabel("")
plt.show()
```

```python
sentence2 = ["This", "model", "focuses", "on", "important", "words"]

attention2 = torch.tensor([0.05, 0.1, 0.25, 0.1, 0.4, 0.1])
attention2 = F.softmax(attention2, dim=0).detach().numpy()

plt.figure(figsize=(10, 2))
sns.heatmap(
    [attention2],
    annot=True,
    xticklabels=sentence2,
    yticklabels=["Attention"],
    cmap="coolwarm"
)

plt.title("Improved Attention Heatmap")
plt.show()
```
